```python
car_missing=pd.read_csv("car-sales-missing-data.csv")
```

```python
car_missing
```

```python
# two ways od fill na
# -> 1) asign the value to that coloumn
car_missing["Odometer"]=car_missing["Odometer"].fillna(car_missing["Odometer"].mean())
car_missing
```

```python
#-> use inplace= True
Odometer_mean= car_missing["Odometer"].mean()
car_missing.fillna({"Odometer":Odometer_mean},inplace=True)
```

```python
# Dropna -> drops the row if it contains null values
car_missing.dropna(inplace=True)
car_missing
```

```python
# exporting the dropped table with no null values
car_missing_dropped=car_missing.dropna()
car_missing_dropped.to_csv("car_missing_dropped.csv", index=False)
```

```python
car_missing_dropped
```

```python
seats= pd.Series([5,5,5,5,5])
# adding new column
car_missing["seats"]= seats
car_missing
```

```python
# -> filling the missing values
car_missing["seats"]=car_missing["seats"].fillna(5)
car_missing.fillna({"seats":5},inplace=True)
car_missing
```

```python
# column from python list
milage=[70,80,60,70,89,90,98,78,56,78]
car_missing["milage"] = milage
car_missing
```

```python
# deriving a column from existing col
car_missing["kms"]=car_missing["Odometer"]/100 * car_missing["milage"]
car_missing
```

```python
# changing the numbers into single decimal
car_missing["Odometer"]=car_missing["Odometer"].map(lambda x: f"{x:g}")
```

```python
car_missing
```

```python
# changing datatypes of a column
car_missing["kms"]=car_missing["kms"].astype(int)
car_missing
```

```python
# creating a column with single value
car_missing["Doors"]=4
car_missing
```

```python
car_missing["verified"]=True
car_missing
```

```python
# drop requires a axis (col -> axis)
# axis 1 -> column 
car_missing=car_missing.drop("verified", axis=1)
car_missing
```

```python
# shuffle the data for randomization
# frac = 1-> 100% , 0.5->50%
car_shuffle=car_missing.sample(frac=1)
```

```python
car_shuffle
```

```python
#reset the shudffled index
#drop removes the indzx
car_shuffle.reset_index(drop=True)
```

```python
# only select 20% data for large datasets
car_shuffle.sample(frac=0.2)
```
