```python
import pandas as pd
```

```python
movie=pd.read_csv("tmdb_5000_movies.csv")
credits=pd.read_csv("tmdb_5000_credits.csv")
```

```python
# mergiing both of them based on movie id

movie=movie.merge(credits, left_on = 'id', right_on='movie_id')
```

```python
movie.columns
```

```python
# getting useful cols for creating a tag

movie = movie[['movie_id','title_x','overview','genres','keywords','cast','crew']]
```

```python
movie.columns
```

```python
movie.rename(columns={"title_x":"title"},inplace=True)
```

```python
movie.columns
```

```python
movie.dropna(inplace=True)
```

```python
movie.isnull().sum()
```

```python
type(movie['genres'][0])
```

```python
#converting str info into list to access
import ast

movie['genres'] = movie['genres'].apply(ast.literal_eval)
movie['keywords'] = movie['keywords'].apply(ast.literal_eval)
movie['cast'] = movie['cast'].apply(ast.literal_eval)
movie['crew'] = movie['crew'].apply(ast.literal_eval)
```

```python
type(movie["genres"][0])
```

```python
#only taking out names of genres
def get_genres(obj):
    genres=[]
    for i in obj:
        genres.append(i['name'])
    return genres
```

```python
movie["genres"]
```

```python
movie['genres'] = movie['genres'].apply(get_genres)
movie['keywords'] = movie['keywords'].apply(get_genres)
```

```python
movie["genres"].head()
```

```python
movie["keywords"].head()
```

```python
movie["cast"][0]
```

```python
#top 3 actors of cast
def top_actors(obj):
    actors = []
    c=0
    for i in obj:
        if c !=3:
            actors.append(i["name"])
            c+=1
        else:
            break
    return actors
```

```python
movie["cast"]=movie["cast"].apply(top_actors)
```

```python
movie["cast"][0]
```

```python
# director
def director(obj):
    l=[]
    for i in obj:
        if i['job']=='Director':
            l.append(i["name"])
            break
    return l
```

```python
movie['crew']=movie['crew'].apply(director)
```

```python
movie["crew"]
```

```python
# removing spaces
def remove_spaces(obj):
    L = []
    for i in obj:
        L.append(i.replace(" ", ""))
    return L
```

```python
movie['genres'] = movie['genres'].apply(remove_spaces)
movie['keywords'] = movie['keywords'].apply(remove_spaces)
movie['cast'] = movie['cast'].apply(remove_spaces)
movie['crew'] = movie['crew'].apply(remove_spaces)
```

```python
movie["cast"][0]
```

```python
movie['tags'] = (movie['genres'] + movie['keywords'] + movie['cast'] + movie['crew'])
```

```python
movie["tags"][0]
```

```python
new_df = movie[['movie_id', 'title', 'tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df
```

```python
new_df["tags"][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv=CountVectorizer(max_features=5000)
cv.fit(new_df["tags"])
vectors= cv.transform(new_df["tags"]).toarray()
```

```python
vectors.shape
```

```python
vectors[0].sum()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity =cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
movie_index =new_df[new_df["title"]=="Avatar"].index[0]
print(movie_index)
```

```python
distance = similarity[movie_index]

```

```python
# sorting the distances in high to low to get most similar movie
mov_list=sorted(list(enumerate(distance)),reverse=True, key = lambda x : x[1])
```

```python
mov_list
```

```python
for i in mov_list[1:6]:
    print(new_df.iloc[i[0]].title)
```

```python
new_df.iloc[47].title
```

```python
new_df["title"] = new_df["title"].str.replace("-", "", regex=False)
```

```python
# func to recommmend
import numpy as np
def recommend(movie):
    if movie.lower() not in new_df["title"].str.lower().values:
        print("MOvie not found")
        return
    mov_index = new_df[new_df["title"].str.lower() == movie.lower()].index[0]
    distance = similarity[mov_index]
    mov_list = np.argsort(distance)[::-1]

    for i in mov_list[1:6]:
        print(new_df.iloc[i].title)
```

```python
recommend("men in black")
```
